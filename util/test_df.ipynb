```python
import pandas as pd
import numpy as np
```

```python
path = "../../data/weatherHistory.csv"
df = pd.read_csv(path)
```

```python
df[:10]
```

```python
temp_df = df.drop(axis=1, labels=['Formatted Date', 'Summary', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Loud Cover', 'Pressure (millibars)', 'Daily Summary', 'Precip Type'])
```

```python
temp_df[:10]
```

```python
temp_df = temp_df.rename(columns={"Temperature (C)": "real", "Apparent Temperature (C)": "predicted"})
```

```python
temp_df[:10]
```

```python
def get_biases(pred, real):
    return [abs(pr - rl) for pr, rl in zip(pred, real)]
```

```python
THRESH = 3
biases = get_biases(temp_df['predicted'], temp_df['real'])
result = [0 if bias >= THRESH else 1 for bias in biases]
```

```python
result[:10]
```

```python
sum(result)
```
